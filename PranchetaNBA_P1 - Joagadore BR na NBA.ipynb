# Gráficos

```python
import plotly.express as px
```

```python
fig = px.bar(DataFrameJogadores, x='Temporadas', y='Jogador', title='Quantidade de Temporadas')
fig.update_xaxes(tickvals=list(range(0, 20, 1)))
fig.show()
```

```python
df_top3 = DataFrameJogadores.query("Temporadas >= Temporadas.nlargest(3).iloc[-1]")
```

```python
df_top3
```

```python
fig_top3_temp = px.bar(df_top3, x='Temporadas', y='Jogador', title='Quantidade de Temporadas')
fig_top3_temp.update_xaxes(tickvals=list(range(0, 20, 1)))
fig_top3_temp.show()
```

```python
fig_top3_games = px.bar(df_top3, x='Jogos', y='Jogador', text='Jogos', title='Quantidade de Jogos')
fig_top3_games.update_xaxes(tickvals=list(range(0, 1000, 50)))
fig_top3_games.show()
```

```python
fig_top3_min = px.bar(df_top3, x='Minutos Jogados', y='Jogador',text='Minutos Jogados', title='Quantidade de Minutos')
fig_top3_min.update_xaxes(tickvals=list(range(0, 30000, 2500)))
fig_top3_min.show()
```

```python
fig_top3_pts = px.bar(df_top3, x='Total Pontos', y='Jogador',text='Total Pontos', title='Quantidade Total de Pontos')
fig_top3_pts.update_xaxes(tickvals=list(range(0, 11000, 500)))
fig_top3_pts.show()
```

```python
fig_top3_pts_3p = px.bar(df_top3, x='Jogador', y='Total Pontos', color='Cestas 3p Convertidas',text='Total Pontos',  title='Quantidade Total de Pontos e cestas de 3 pontos')
fig_top3_pts_3p.show()
```

```python
fig_top3_pts_2p = px.bar(df_top3, x='Jogador', y='Total Pontos', color='Cestas 2p Convertidas',text='Total Pontos',  title='Quantidade Total de Pontos e cestas de 2 pontos')
fig_top3_pts_2p.show()
```

```python
fig_top3_pts_ll = px.bar(df_top3, x='Jogador', y='Total Pontos', color='Lances Livres Convertidos', text='Total Pontos', title='Quantidade Total de Pontos e Lances Livres Convertidos')
fig_top3_pts_ll.show()
```
