```python
import pandas as pd
import matplotlib.pyplot as plt
from nba_api.stats.static import players
from nba_api.stats.static import teams
from nba_api.stats.endpoints import playergamelog
from nba_api.stats.library.parameters import SeasonAll
```

# Definindo VÃ¡riaveis

```python
teams = teams.get_teams()
```

```python
players_dict = players.get_players()
```

```python
#Visualizando dados dos jogadores
players_dict
```

```python
#Visualizando dados dos times
teams
```

# Criando Data Frames

```python
#Dataframe dos jogadores
df_players = pd.DataFrame(players_dict)
```

```python
#Data Frame dos times
df_teams = pd.DataFrame(teams)
```

```python
df_teams
```

```python
df_players.head()
```

# Selecionando Atletas

```python
#Selecionando Atletas = Curry e Kobe
players = ["Stephen Curry", "Kobe Bryant"]
```

```python
#Filtrando jogadores pelo nome completo
df_id = df_players.query('full_name == @players')
```

```python
df_id
```

```python
df_Kobe = playergamelog.PlayerGameLog(player_id=df_id.iat[0,0], season = SeasonAll.all).get_data_frames()[0]
```

```python
df_Curry = playergamelog.PlayerGameLog(player_id=df_id.iat[1,0], season = SeasonAll.all).get_data_frames()[0]
```

```python
df_Kobe.head()
```

```python
df_Curry.head()
```

# Tratando Data Frame

```python
nomes_colunas_curry = df_Curry.columns.tolist()
print(nomes_colunas_curry)
```

```python
#Eliminando Colunas que nÃ£o serÃ£o uteis
df_Curry = df_Curry.drop(["Player_ID", "Game_ID", "GAME_DATE", "MATCHUP", "WL", "MIN", "FGM", "FGA","FG3M", "FG3A", "FG3_PCT", "FTM", "FTA","OREB", "DREB", "PF", "PLUS_MINUS", "VIDEO_AVAILABLE"], axis = 'columns')
```

```python
df_Curry
```

```python
#Eliminando Colunas que nÃ£o serÃ£o uteis
df_Kobe = df_Kobe.drop(["Player_ID", "Game_ID", "GAME_DATE", "MATCHUP", "WL", "MIN", "FGM", "FGA","FG3M", "FG3A", "FG3_PCT", "FTM", "FTA","OREB", "DREB", "PF", "PLUS_MINUS", "VIDEO_AVAILABLE"], axis = 'columns')
df_Kobe
```

```python
#Tratando coluna Season_ID, para mostrar o ano correto
df_Kobe['SEASON_ID'] = df_Kobe['SEASON_ID'].str[1:5]
df_Curry['SEASON_ID'] = df_Curry['SEASON_ID'].str[1:5]
```

```python
df_Kobe
```

```python
df_Curry
```

```python
#Agrupando dados de acordo com o ano
df_Curry = df_Curry.groupby('SEASON_ID').mean()
df_Kobe = df_Kobe.groupby('SEASON_ID').mean()
```

```python
df_Kobe
```

```python
df_Kobe.reset_index(inplace=True)
df_Curry.reset_index(inplace=True)
```

# Criando GrÃ¡ficos

```python
plt.figure(figsize=(14,10))
plt.rcParams['xtick.labelsize'] = 10
plt.xticks(rotation=60)
plt.rcParams['ytick.labelsize'] = 12
plt.xlabel('Temporadas', size = 12)
plt.ylabel('MÃ©dia de Rebotes por Jogo', size = 12)
plt.scatter(df_Kobe['SEASON_ID'], df_Kobe['REB'], marker = '.', alpha = 0.8, s = 100, color = 'orange', label = 'Kobe REB' )
plt.scatter(df_Curry['SEASON_ID'], df_Curry['REB'], marker = '.', alpha = 0.8, s = 100, color = 'blue', label = 'Curry REB' )
plt.plot(df_Kobe['SEASON_ID'], df_Kobe['REB'], color = 'orange',lw=1 )
plt.plot(df_Curry['SEASON_ID'], df_Curry['REB'],color = 'blue',lw=1 )
plt.axhline(df_Kobe['REB'].mean(), 0,1, color = 'orange', **{'ls':'--', 'lw': 0.8})
plt.axhline(df_Curry['REB'].mean(), 0,1, color = 'blue', **{'ls':'--', 'lw': 0.8})
plt.text(.4,1.8,"InÃ­cio Carreira de Kobe", bbox = {'facecolor': 'gold', 'alpha': 0.5})
plt.text(8,4.4,"InÃ­cio Carreira de Curry", bbox = {'facecolor': 'lightblue', 'alpha': 0.5})
plt.text(19.4,3.7,"Fim Carreira de Kobe", bbox = {'facecolor': 'gold', 'alpha': 0.5})
plt.annotate('mÃ©dia Kobe', xy=(-1,5.22), color='orange')
plt.annotate('mÃ©dia Curry', xy=(-1,4.78), color='blue')
plt.legend()
plt.grid(axis = 'x', color='grey', linewidth = '0.15')
plt.savefig('PranchetaNBA-P4-REB.png', format='png')
```
