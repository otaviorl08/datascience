```python
#importando biblioteca Pandas
import pandas as pd
import numpy as np
```

```python
#Criando conteudo do Data Frame
data = {'Cidade': ['São Paulo', 'Rio de Janeiro', 'João Pessoa', 'Porto Alegre','Manaus'],
        'Total de Doentes': [19045, None, 8901, 13467, 14569],
        'Total de Curados': [12345.0, 10890.0, 5467.0, None, None],
        'Total de Pessoas que fizeram o tratamento': [None, 13549.0, 6578.0, 3467.0, None]}
```

```python
#Criando Data Frame
dados = pd.DataFrame(data)
#Exibindo Data Frame
dados.round()
```

```python
#Mostrando onde tem dado nulo no Data Frame com o comando isnull()
#A posção que tiver dado nulo mostrará: True
dados.isnull()
```

```python
#Calculando a média de cada coluna
mediadoentes = dados['Total de Doentes'].mean()
mediacurados = dados['Total de Curados'].mean()
mediatratamento = dados['Total de Pessoas que fizeram o tratamento'].mean()
```

```python
#Preenchendo os espaços com dados nulos com a média de cada coluna
dados['Total de Doentes'] = dados['Total de Doentes'].fillna(mediadoentes)
dados['Total de Curados'] = dados['Total de Curados'].fillna(mediacurados)
dados['Total de Pessoas que fizeram o tratamento'] = dados['Total de Pessoas que fizeram o tratamento'].fillna(mediatratamento)

dados.round()
```

```python
#Calculando as principais medidas de disperção e posição das colunas

novamediadoentes = dados['Total de Doentes'].mean()
novamediacurados = dados['Total de Curados'].mean()
novamediatratamento = dados['Total de Pessoas que fizeram o tratamento'].mean()

medianadoentes = dados['Total de Doentes'].median()
medianacurados = dados['Total de Curados'].median()
medianatratamento = dados['Total de Pessoas que fizeram o tratamento'].median()

varianciadoentes = dados['Total de Doentes'].var()
varianciacurados = dados['Total de Curados'].var()
varianciatratamento = dados['Total de Pessoas que fizeram o tratamento'].var()

desviopadraodoentes = dados['Total de Doentes'].std()
desviopadraocurados = dados['Total de Curados'].std()
desviopadraotratamento = dados['Total de Pessoas que fizeram o tratamento'].std()
```

```python
#criando dataset com medidas
dadosexplorados = {'Amostra': ['Pessoas Doentes', 'Pessoas Curadas', 'Pessoas que fizeram tratamento'],
                   'Média': [novamediadoentes, novamediacurados, novamediatratamento],
                   'Mediana': [medianadoentes, medianacurados, medianatratamento],
                   'Variância': [varianciadoentes, varianciacurados, varianciatratamentos],
                   'Desvio Padrão': [desviopadraodoentes, desviopadraocurados, desviopadraotratamento]}
```

```python
#Criando novo dataframe das medidas de posição e disperção
dfdadosexplorados = pd.DataFrame(dadosexplorados)
dfdadosexplorados.round()
```
