# Importando Bibliotecas e Realizando WebScraping

```python
#Importando Bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Scraping URL Lebron James
urlLJ = 'https://www.espn.com.br/nba/jogador/estatisticas/_/id/1966/lebron-james'
```

```python
#Scraping URL Lebron James
urlMJ = 'https://www.espn.com.br/nba/jogador/estatisticas/_/id/1035/michael-jordan'
```

# Criando Data Frame Lebron James

```python
dadosLJ = pd.read_html(urlLJ)
dadosLJ
```

```python
#Separando Colunas informativas de colunas de dados
```

```python
dadosLJ[0].head(10)
```

```python
dadosLJ[1].head(10)
```

```python
#Removendo colunas que não serão utilizadas
DataFrameLJ1 = dadosLJ[1].drop(["GS","FG","FG%","3PT","3P%","FT","FT%","OR","DR","REB","AST","BLK","STL","PF","TO"], axis = 'columns')
DataFrameLJ1.head(10)
```

```python
dadosLJ[3].head(5)
```

```python
DataFrameLJ2 = dadosLJ[3].drop(["OR","DR","BLK","STL","PF","TO"], axis = 'columns')
DataFrameLJ2
```

```python
DataFrameLJ2.rename(columns={'PTS':'Points'}, inplace=True)
DataFrameLJ2.head(10)
```

```python
dadosLJ[5].head(3)
```

```python
DataFrameLJ3 = dadosLJ[5].drop(["DD2","DQ","EJECT","TECH","FLAG","AST/TO","STL/TO","RAT","SC-EFF","SH-EFF"], axis = 'columns')
DataFrameLJ3
```

```python
DataFrameLJ0 = dadosLJ[0]
DataFrameLJ0
```

```python
DataFrameLJfinal = pd.concat([DataFrameLJ0, DataFrameLJ1, DataFrameLJ2, DataFrameLJ3], axis = 1)
DataFrameLJfinal.head(10)
```

```python
DataFrameLJfinal.rename(columns={'season': 'Season', 'J': 'Games','MIN': 'Minutes', 'PTS': 'Points Average', 'FG':'Shots taken - Shots attempted', 'FG%':'Percentage taken', 'FT': 'Free Throw', 'V':'Assistances', 'TD3':'Triple-Double'}, inplace=True)
DataFrameLJfinal
```

```python
DataFrameLJfinal = DataFrameLJfinal.replace('Carreira', 'Total')
```

```python
DataFrameLJfinal["Time"].fillna(" ", inplace = True)
DataFrameLJfinal.head()
```

```python
n = len(DataFrameLJfinal)
newcolumn = np.arange(1, n + 1)
newcolumnname = 'Ano da Carreira'
posicao = 0

DataFrameLJfinal.insert(posicao, newcolumnname, newcolumn)
```

```python
#Visualizando o final do DF pra ver até que ano LJ jogou/esta jogando
DataFrameLJfinal.tail()
```

```python
dfsemtotalLJ = DataFrameLJfinal.drop(20, axis='index')
dfsemtotalLJ.tail()
```

#Criando Gráficos de comparação durante a carreira

```python
#Criando Figura
plt.figure(figsize=(12, 10))
#Criando Linhas
plt.plot(dfsemtotalMJ['Ano da Carreira'], dfsemtotalMJ['Games'], marker='o', color='red', label='Michael Jordan', lw=2)
plt.plot(dfsemtotalLJ['Ano da Carreira'], dfsemtotalLJ['Games'], marker='o', color='gold', label='Lebron James', lw=2)
#Alterando Intervalo e Periodo dos eixos
intervaloX = np.arange(0, 22, 1)
intervaloY = np.arange(0, 100, 5)
plt.xticks(intervaloX)
plt.yticks(intervaloY)
#Grade
plt.grid(True)
#Legenda
plt.legend()
#Titulo dos Eixos
plt.xlabel('Temporada da Carreira')
plt.ylabel('Jogos por temporada')
#Titulo do Grafico
plt.title('Quantidade de jogos por temporada')
#Salva o Gráfico
plt.savefig('jogos_por_temporada.png')
#Mostra o Grafico
plt.show()
```

```python
linhatotalMJ['Atleta'] = ['Michael Jordan']
linhatotalMJ
```

```python
DataFrameTotal = pd.merge(linhatotalLJ, linhatotalMJ, how='outer')
DataFrameTotal
```

```python
ColunaAtletas = DataFrameTotal.pop("Atleta")
```

```python
 DataFrameTotal.insert(0, 'Atleta', ColunaAtletas)
```

```python
DataFrameTotal
```

```python
ColunasRemover = ['Ano da Carreira', 'Season', 'Time']
```

```python
DataFrameTotal = DataFrameTotal.drop(ColunasRemover, axis=1)
```

```python
DataFrameTotal
```

```python
#Criando Figura
plt.figure(figsize=(6, 6))
#Criando Barras
cores = ['gold', 'r']
plt.bar(DataFrameTotal['Atleta'], DataFrameTotal['Games'], color = cores, edgecolor='black')
#Titulo dos Eixos
plt.xlabel('Jogador')
plt.ylabel('Jogos na Carreira')
plt.title('Quantidade de Jogos na NBA')
#Alterando Intervalo e Periodo dos eixos
intervaloY = np.arange(0, 1600, 100)
plt.yticks(intervaloY)
#Grade
plt.grid(axis='y', which='major')
#Salva o Gráfico
plt.savefig('quantidade_jogos_total.png')
# Exibindo o gráfico
plt.tight_layout()
plt.show()
```

```python
#Criando Figura
plt.figure(figsize=(6, 6))
#Criando Barras
cores = ['gold', 'r']
plt.bar(DataFrameTotal['Atleta'], DataFrameTotal['Points'], color = cores, edgecolor='black')
#Titulo dos Eixos
plt.xlabel('Jogador')
plt.ylabel('Pontos na Carreira')
plt.title('Quantidade de Pontos na NBA')
#Alterando Intervalo e Periodo dos eixos
intervaloY = np.arange(0, 40000, 2000)
plt.yticks(intervaloY)
#Grade
plt.grid(axis='y', which='major')
#Salva o Gráfico
plt.savefig('quantidade_pontos_total.png')
# Exibindo o gráfico
plt.tight_layout()
plt.show()
```
