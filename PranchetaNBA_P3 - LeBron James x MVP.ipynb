#Importando Biblioteca

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
import numpy as np
```

#Importando Arquivo CSV e criando Data Frame

```python
arquivomvp = 'datamvp.csv'
```

```python
dataframemvp = pd.read_csv(arquivomvp)
```

```python
dataframemvp
```

```python
arquivolebron = 'datalebron.csv'
```

```python
dataframelebron = pd.read_csv(arquivolebron)
dataframelebron
```

#Excluindo colunas indesejadas

```python
dataframemvp = dataframemvp.drop(columns=['TRB', 'BLK', 'FG%', 'STL', '-9999'])
dataframemvp
```

```python
dataframelebron = dataframelebron.drop(columns=['2P%'])
dataframelebron
```

```python
dataframelebron = dataframelebron.drop(dataframelebron.index[-1])
dataframelebron
```

#Renomeando Colunas

```python
dataframemvp.rename(columns={'Age':'Idade-MVP', 'Tm': 'Time-MVP',  'G': 'Jogos-MVP', 'MP': 'Min por jogo-MVP', 'PTS':'Pontos por jogo-MVP', 'AST':'Assistencia por jogo-MVP', '3P%':'Porcentagem 3P por jogo-MVP', 'FT%':'Porcentagem lance livre por jogo-MVP'}, inplace=True)
dataframemvp
```

```python
dataframelebron.rename(columns={ 'Age':'Idade-Lebron', 'Tm':'Time-Lebron', 'G': 'Jogos-Lebron', 'MP': 'Min por jogo-Lebron', 'PTS':'Pontos por jogo-Lebron', 'AST':'Assistencia por jogo-Lebron', '3P%':'Porcentagem 3P por jogo-Lebron', 'FT%':'Porcentagem lance livre por jogo-Lebron'}, inplace=True)
dataframelebron
```

#Configurando Data Frame final

```python
dataframeunificado = pd.merge(dataframemvp, dataframelebron, on='Season')
dataframeunificado
```

```python
dataframeunificado = dataframeunificado[::-1]
dataframeunificado
```

```python
n = len(dataframeunificado)
newcolumn = np.arange(1, n+1)
newcolumnname = 'Temporada'
posicao = 0

dataframeunificado.insert(posicao, newcolumnname, newcolumn)
```

```python
dataframeunificado
```

#Criando Grficos

```python
plt.figure(figsize=(10, 8))
plt.plot(dataframeunificado['Season'], dataframeunificado['Jogos-Lebron'], marker='*', label='Lebron', lw=1)
plt.plot(dataframeunificado['Season'], dataframeunificado['Jogos-MVP'], marker='.', label='MVP', lw=1)
intervaloX = np.arange(0, 22, 1)
intervaloY = np.arange(44, 84, 1)
plt.xticks(intervaloX, fontsize=8, rotation=90)
plt.yticks(intervaloY, fontsize=8)
plt.grid(True)
plt.legend()
plt.xlabel('Temporada da Carreira')
plt.ylabel('Jogos por temporada')
plt.title('Quantidade de jogos por temporada')
plt.savefig('qtde_jogos_temp.png')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.plot(dataframeunificado['Season'], dataframeunificado['Pontos por jogo-Lebron'], marker='*', label='Lebron', lw=1)
plt.plot(dataframeunificado['Season'], dataframeunificado['Pontos por jogo-MVP'], marker='.', label='MVP', lw=1)
intervaloX = np.arange(0, 22, 1)
intervaloY = np.arange(15, 35, 1)
plt.xticks(intervaloX, fontsize=8, rotation=90)
plt.yticks(intervaloY, fontsize=8)
plt.grid(True)
plt.legend()
plt.xlabel('Temporada da Carreira')
plt.ylabel('Pontos por jogo')
plt.title('Mdia pontos por jogo')
plt.savefig('media_pontos_por_jogo.png')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.plot(dataframeunificado['Season'], dataframeunificado['Idade-Lebron'], marker='*', label='Lebron', lw=1)
plt.plot(dataframeunificado['Season'], dataframeunificado['Idade-MVP'], marker='.', label='MVP', lw=1)
intervaloX = np.arange(0, 22, 1)
intervaloY = np.arange(18, 42, 1)
plt.xticks(intervaloX, fontsize=8, rotation=90)
plt.yticks(intervaloY, fontsize=8)
plt.grid(True)
plt.legend()
plt.xlabel('Temporada da Carreira')
plt.ylabel('Idade do MVP')
plt.title('Idade do MVP')
plt.savefig('idade_mvp.png')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.plot(dataframeunificado['Season'], dataframeunificado['Porcentagem 3P por jogo-Lebron'], marker='*', label='Lebron', lw=1)
plt.plot(dataframeunificado['Season'], dataframeunificado['Porcentagem 3P por jogo-MVP'], marker='.', label='MVP', lw=1)
intervaloX = np.arange(0, 22, 1)
intervaloY = np.arange(0, 1, 0.1)
plt.xticks(intervaloX, fontsize=8, rotation=90)
plt.yticks(intervaloY, fontsize=8)
plt.grid(True)
plt.legend()
plt.xlabel('Temporada da Carreira')
plt.ylabel('% 3 Pontos por Jogo')
plt.title('% 3 Pontos por Jogo')
plt.savefig('porc_3pontos_por_jogo.png')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.plot(dataframeunificado['Season'], dataframeunificado['Assistencia por jogo-Lebron'], marker='*', label='Lebron', lw=1)
plt.plot(dataframeunificado['Season'], dataframeunificado['Assistencia por jogo-MVP'], marker='.', label='MVP', lw=1)
intervaloX = np.arange(0, 22, 1)
intervaloY = np.arange(0, 15, 1)
plt.xticks(intervaloX, fontsize=8, rotation=90)
plt.yticks(intervaloY, fontsize=8)
plt.grid(True)
plt.legend()
plt.xlabel('Temporada da Carreira')
plt.ylabel('Mdia Assistncia por Jogo')
plt.title('Mdia Assistncia por Jogo')
plt.savefig('media_ass_por_jogo.png')
plt.show()
```
